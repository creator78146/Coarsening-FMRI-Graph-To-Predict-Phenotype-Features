```python
import numpy as np
import matplotlib.pyplot as plt

g = np.load("../graphs/subject_0_graph.npy")

plt.figure(figsize=(6,6))
plt.imshow(g, cmap='hot')
plt.colorbar()
plt.title("Brain Connectivity (Adjacency Matrix)")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
# Load graph
g = np.load("../graphs/subject_0_graph.npy")
# Create graph object
G = nx.from_numpy_array(g)
plt.figure(figsize=(8,8))
nx.draw(G, node_size=20, edge_color='yellow')
plt.title("Brain Graph Network")
plt.show()
```

```python
threshold = 0.5

g_filtered = (g > threshold) * g

G = nx.from_numpy_array(g_filtered)

plt.figure(figsize=(8,8))
nx.draw(G, node_size=30)
plt.title("Filtered Brain Graph")
plt.show()
```
