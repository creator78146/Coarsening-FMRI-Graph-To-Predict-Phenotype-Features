```python
import torch
from torch_geometric.data import Data
labels = df['DX_GROUP'].replace({1:1, 2:0}).reset_index(drop=True)

dataset = []

for i in range(len(graphs)):
    adj = graphs[i]

    edge_index = np.array(np.nonzero(adj))
    edge_index = torch.tensor(edge_index, dtype=torch.long)

    x = torch.eye(adj.shape[0], dtype=torch.float)

    y = torch.tensor([labels[i]], dtype=torch.long)

    data_obj = Data(x=x, edge_index=edge_index, y=y)
    dataset.append(data_obj)
```

```python
from torch_geometric.nn import GCNConv, global_mean_pool
import torch.nn.functional as F
import torch

class GCN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = GCNConv(30, 64)
        self.conv2 = GCNConv(64, 2)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        x = self.conv1(x, edge_index)
        x = F.relu(x)

        x = self.conv2(x, edge_index)
        
        batch = torch.zeros(x.size(0), dtype=torch.long)

        x = global_mean_pool(x, batch)

        return x
```

```python
labels = df['DX_GROUP'].replace({1:1, 2:0}).reset_index(drop=True)

dataset = []

for i in range(len(graphs)):
    adj = graphs[i]

    edge_index = np.array(np.nonzero(adj))
    edge_index = torch.tensor(edge_index, dtype=torch.long)

    x = torch.eye(adj.shape[0], dtype=torch.float)

    y = torch.tensor([labels[i]], dtype=torch.long)

    data_obj = Data(x=x, edge_index=edge_index, y=y)
    dataset.append(data_obj)
```

## from here we have perform train test slpit to train our model and evaluate performance matrices

```python
def get_predictions(model, data_list):
    model.eval()
    
    y_true = []
    y_pred = []

    for data in data_list:
        out = model(data)
        pred = out.argmax().item()

        y_pred.append(pred)
        y_true.append(data.y.item())

    return y_true, y_pred
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(
    dataset, 
    test_size=0.2, 
    random_state=42
)

print(len(train_data), len(test_data))
```

```python
model = GCN()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(30):
    model.train()
    total_loss = 0

    for data in train_data:   # 🔥 ONLY train_data
        optimizer.zero_grad()

        out = model(data)
        loss = F.cross_entropy(out, data.y)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch}, Loss: {total_loss:.4f}")
```

```python
y_true, y_pred = get_predictions(model, test_data)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_true, y_pred))
print("Precision:", precision_score(y_true, y_pred))
print("Recall:", recall_score(y_true, y_pred))
print("F1 Score:", f1_score(y_true, y_pred))

print("\nConfusion Matrix:\n", confusion_matrix(y_true, y_pred))

print("\nFull Report:\n", classification_report(y_true, y_pred))
```
