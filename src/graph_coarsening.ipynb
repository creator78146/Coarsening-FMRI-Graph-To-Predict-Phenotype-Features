```python
import numpy as np
import os
from sklearn.cluster import KMeans

os.makedirs("../coarse_graphs", exist_ok=True)

for i in range(20):
    g = np.load(f"../graphs/subject_{i}_graph.npy")
    
    kmeans = KMeans(n_clusters=30, random_state=0)
    labels = kmeans.fit_predict(g)
    
    coarse_graph = np.zeros((30, 30))
    
    for a in range(len(labels)):
        for b in range(len(labels)):
            c1 = labels[a]
            c2 = labels[b]
            coarse_graph[c1][c2] += g[a][b]
    
    coarse_graph = coarse_graph / np.max(coarse_graph)
    
    np.save(f"../coarse_graphs/subject_{i}.npy", coarse_graph)

print("Coarsening completed!")
```

```python
import matplotlib.pyplot as plt

plt.imshow(coarse_graph, cmap='hot')
plt.colorbar()
plt.title("Coarsened Graph")
plt.show()
```
