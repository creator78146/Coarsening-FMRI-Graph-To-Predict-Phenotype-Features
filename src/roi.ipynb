```python
from nilearn.datasets import fetch_abide_pcp

data = fetch_abide_pcp(
    data_dir="../data/",
    pipeline='cpac',
    band_pass_filtering=True,
    global_signal_regression=True,
    n_subjects=20
)

print(len(data.func_preproc))
```

```python
from nilearn.input_data import NiftiLabelsMasker
import numpy as np
import os

# Create folder
os.makedirs("processed_data", exist_ok=True)

# Load atlas
atlas_path = "../data/atlas/AAL3v1.nii"

masker = NiftiLabelsMasker(
    labels_img=atlas_path,
    standardize=True
)

# Process all subjects
for i, file in enumerate(data.func_preproc):
    print(f"Processing subject {i}...")

    ts = masker.fit_transform(file)
    ts = ts.T

    np.save(f"processed_data/subject_{i}.npy", ts)

print("✅ ROI extraction completed!")
```
