```python
import os
import numpy as np

os.makedirs("../graphs", exist_ok=True)

for i in range(20):
    ts = np.load(f"../processed_data/subject_{i}.npy")
    
    corr_matrix = np.corrcoef(ts)
    
    # FIX
    corr_matrix = np.nan_to_num(corr_matrix)

    np.save(f"../graphs/subject_{i}_graph.npy", corr_matrix)

print("✅ graphs created!")
```
